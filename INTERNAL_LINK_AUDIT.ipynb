```python
##Library to processing the dataset
import pandas as pd
```

```python
#Pulling the dataset extracted form Screaming Frog
df = pd.read_csv("internal_all.csv")
list(df.columns)
#selecting the necessary columns
selected_columns = df[["Address", 'Ahrefs URL Rating - Exact', 'Ahrefs RefDomains - Exact', 'Ahrefs DoFollow - Exact',
                       'Crawl Depth','Link Score','Inlinks',
                       'GA Goal Completions All',
                       'Clicks', 'Impressions', 'CTR', 'Position',
                       'Majestic Referring Domains - URL (Exact URL)', 'Majestic Trust Flow - URL (Exact URL)']]
#making a dataset copy
new_df = selected_columns.copy()
#filling NA with Zeros
new_df=new_df.fillna(0)
```

```python
#Applying the conditions
#Root Domains (Below 30) and Trials (Over 10) = Needs External Links
a=(new_df['Ahrefs RefDomains - Exact']<30)&(new_df['GA Goal Completions All']>10)
#Link Score (Below 30) and Trials (Over 10) = Needs Internal Links
b=(new_df['Link Score']<30)&(new_df['GA Goal Completions All']>10)
#Root Domains (Over 50), Link Score (Over 30) and Trials (0) = Boost Another Page
c=(new_df['Ahrefs RefDomains - Exact']>50)&(new_df['Link Score']>30)&(new_df['GA Goal Completions All']==0)
#creating the new columns with the processed conditions
new_df["external link boost"]=a
new_df["internal link boost"]=b
new_df["boost another page"]=c
```

```python
#replacing boolean results (ease reading)
mask = new_df.applymap(type) != bool
d = {True: 'Yes', False: 'No'}
new_df = new_df.where(mask, new_df.replace(d))
#Sorting by Best Trial (GA Goals)
new_df=new_df.sort_values(by = ['GA Goal Completions All','CTR'], ascending=False)
#Displaying first 15 urls 
with pd.option_context('display.max_colwidth', None):
  display(new_df.head(15))
```
